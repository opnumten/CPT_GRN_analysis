```r
setwd('C:/Users/zoro/Desktop/Frustration_paper/data_analysis/dyn_version/result/')
library(locfdr)
library(reticulate)

np <-import("numpy")


F_sparse<-np$load('dyn_EMT_F.npy')
F_data<-F_sparse[,3]
# F_data<-abs(F_sparse[,3])
print(dim(F_sparse))
w <- locfdr(F_data,bre=max(ceiling(1e-3*length(F_data)),100),df=10,nulltype=1,plot=1)#type=1,pct=0,,pct0=0.25
#     hist(F_sparse[,3],breaks=100)
F_sparse[,3][which(w$fdr>=0.3)]<-0
hist(F_sparse[,3],breaks=100)
np$save(file.path(getwd(),'sparse_EMT_F.npy'),F_sparse)
print(length(which(w$fdr<=0.3)))
```
